```
import pandas as pd
df = pd.read_csv('hit3dummy.csv')
df
```

```
def findUnqualified(df):
  unqualified = []
  for row in df.iterrows():
    row = row[1]
    neg1 = row["NegImg1"]
    neg2 = row["NegImg2"]
    worker = row["WorkerID"]
    for i in range(1,9):
      rank = row["Rank" + str(i)]
      if rank == neg1 or rank == neg2 and worker not in unqualified:
        unqualified.append(row["WorkerID"])
  return(unqualified)
    
```

```
print(findUnqualified(df))
```
